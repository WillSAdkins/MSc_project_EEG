```python
def calc_z_value(mu, std):
 return (mu-25)/std

calc_z_value(33, 3.1)
```

```python
def calc_significance(data):
  for i in range(len(data)):
    if calc_z_value(data[i][0], data[i][1]) > 1.64:
      print(data[i])

calc_significance([(33, 3.8), (28.1, 8.8), (32.9, 5.5), (34.9, 4.2), (34.1, 2.8), (30.3, 2.6), (28.5, 3.6), (31.2, 1.7), (38.9, 5.6), (28.8, 7.0), (29.0, 3.6), (28.2, 3.8), (28.3, 2.7), (31.1, 6.1)])
```

```python
def calc_significance_binary(data):
  for i in range(len(data)):
    if (data[i][0]-50)/data[i][1] > 1.64:
      print(data[i])

calc_significance_binary([(56.6, 5.7), (54, 2), (58.3, 7.9), (61.1, 4.3), (62.3, 2.4), (56.6, 4.0), (61.4, 6.9), (60.9, 7.9), (63.2, 2.7), (54.5, 3.3), (56.8, 4.9), (62.7, 8.0)])
```
